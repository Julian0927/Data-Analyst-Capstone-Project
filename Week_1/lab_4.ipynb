## ***Extract information from the given web site***


$(1)$ You will extract the data from the below web site:


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**. It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.

$ \ $

$(2)$ Import the required libraries.


```python
#(1) this module helps in web scrapping.
from bs4 import BeautifulSoup 

#(2) this module helps us to download a web page
import requests  
```

$ \ $ 

$(3)$ Download the webpage at the url.

```python
data  = requests.get(url).text 
```

$ \ $ 

$(4)$ Create a soup object.

```python
soup = BeautifulSoup(data, "html5lib") 
```

$ \ $

$(5)$ Scrape the `Language name` and `annual average salary`.


```python
table = soup.find('table') 
#table
```

image.png

$ \ $ 

$(6)$ Save the scrapped data into a file named ***`popular-languages.csv`***

```python
# Get all rows from the table. An html table row is represented by the tag <tr>
for row in table.find_all('tr'):
  
    # Get all columns in each row.    
    # in html a column is represented by the tag <td>
    cols = row.find_all('td') 
    
    # store the value in column 2 as color_name
    color_name = cols[1].getText()

    # store the value in column 3 as color_code
    color_code = cols[3].getText()

    # show on the screen the results
    print(f"{color_name} ---> {color_code}")

    # space
    print("\n")
```
