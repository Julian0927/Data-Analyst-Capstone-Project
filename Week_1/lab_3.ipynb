## ***Scrape www.ibm.com***

$ \ $

$(1)$ Import the required modules and functions.

```python
#(1) this module helps in web scrapping.
from bs4 import BeautifulSoup 

#(2) this module helps us to download a web page
import requests  
```

$ \ $

$(2)$ Download the contents of the web page.

```python
url = "http://www.ibm.com"
```

```python
# get the contents of the webpage in text format and store in a variable called data
data  = requests.get(url).text 
```

$ \ $

$(3)$ Create a soup object using the class BeautifulSoup.

```python
# create a soup object using the variable 'data'
soup = BeautifulSoup(data, "html5lib")  
```

$ \ $

$(4)$ Scrape all links.

```python
# in html anchor/link is represented by the tag <a>
for link in soup.find_all('a'):  
    print(link.get('href'))
```

$ \ $

$(5)$ Scrape all images.

```python
# in html image is represented by the tag <img>
for link in soup.find_all('img'):
  print(link.get('src'))
```

## ***Scrape data from html tables***

The below url contains a html table with data about colors and color codes.

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/HTMLColorCodes.html"
```

$ \ $

Before proceeding to scrape a web site, you need to examine the contents, and the way data is organized on the website. For this task, we will do the following steps:

$ \ $

$(1)$ Open the above url in your browser and check how many rows and columns are there in the color table.

```python
# get the contents of the webpage in text format and store in a variable called data
data  = requests.get(url).text
```

```python
soup = BeautifulSoup(data,"html5lib")
```

```python
# find a html table in the web page. This is represented by the tag <table>
table = soup.find('table') 
#table
```

```python
# Get all rows from the table. An html table row is represented by the tag <tr>
for row in table.find_all('tr'):
  
    # Get all columns in each row.    
    # in html a column is represented by the tag <td>
    cols = row.find_all('td') 
    
    # store the value in column 2 as color_name
    color_name = cols[2].getText()

    # store the value in column 3 as color_code
    color_code = cols[3].getText()

    # show on the screen the results
    print(f"{color_name} ---> {color_code}")

    # space
    print("\n")
```
