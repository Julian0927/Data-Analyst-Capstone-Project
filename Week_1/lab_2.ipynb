$(1)$ Import the url.

```python
baseurl = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/datasets/githubposting.json"
```

$ \ $

$(2)$ Use the ***get*** function to obtain a request.

```python
response = requests.get(baseurl) 
```

$ \ $

$(3)$ Show the results given using the ***json()*** function.

```python
if response.ok:             
  data = response.json()  
  print(data)  
```

```python
data.keys()
```

$ \ $

$(4)$ Because data is a dictionary and one of its keys is 'technology', then we extract the respective value.

```python
#data['technology']
```

$ \ $

$(5)$ Because data is a dictionary and one of its keys is 'number of job posting', then we extract the respective value.

```python
#data['number of job posting']
```

```python
def get_number_of_jobs(technology):
    number_of_jobs = 0
    technologies = data["technology"]
    number_job_pos = data["number of job posting"]
    for key in technologies:
      A = technologies[key].lower()
      B = technology.lower()
      if A == B:
        number_of_jobs += int(number_job_pos[key])
    return number_of_jobs
```

```python
print(get_number_of_jobs('python'))
```

## ***Store the results in an excel file***


Call the API for all the given technologies above and write the results in an excel spreadsheet. For this task, we have the following steps:

$ \ $

$(1)$ We create a python list of all technologies for which you need to find the number of jobs postings.

```python
technologies = list(set(data["technology"].values()))
technologies
```

$ \ $

$(2)$ Import libraries required to create excel spreadsheet.

```python
!pip3 install openpyxl
from openpyxl import Workbook
```

$ \ $

$(3)$ Create a workbook and select the active worksheet

```python
wb = Workbook() 
ws = wb.active
ws
```

$ \ $

$(4)$ Find the number of jobs postings for each of the technology in the above list. Write the technology name and the number of jobs postings into the excel spreadsheet

```python
tech_list = dict()
for language in technologies:
  tech_list[language] = get_number_of_jobs(language)
tech_list
```

```python
ws.append(['Language', 'Job Postings'])
for key in tech_list:
  ws.append([key,tech_list[key]])
```
