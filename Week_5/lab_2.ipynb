```python
df_age = df[["Age", "Respondent"]].groupby("Age").count().sort_values(by = "Respondent", ascending=False).reset_index()
df_age.head()
```

$(4)$ In this part, we need to describe the following things:

* Capture Respondent Count by Gender, classified by Formal Education Level.

* Visualize as a Stacked bar chart.

* Utilize Bars, Length, Color fields of Stacked bar chart.

* Include Show value labels feature.

* Include a proper Chart title.

```python
df[["Gender"]].value_counts()
```

```python
gender_values = list(df[["Gender"]].value_counts().reset_index()["Gender"].to_dict().values())
gender_values
```

```python
educ_values = list(df[["EdLevel"]].value_counts().reset_index()["EdLevel"].to_dict().values())
educ_values
```

```python
df_edu = df[["Respondent", "Gender", "EdLevel"]].groupby(["Gender", "EdLevel"]).count().sort_values(by = "Gender", ascending=True).reset_index()
df_edu
```

```python
pivot = pd.pivot_table(data=df[["Respondent", "Gender", "EdLevel"]], index=['Gender'], columns=['EdLevel'], values='Respondent')
pivot
```

```python
ax = pivot.plot.bar(stacked=True, figsize=(20,15))
ax.set_title('Education Level Gender', fontsize=20)
ax.set_ylim(0,140000)
```
