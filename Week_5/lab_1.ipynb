$(3)$ In this part, we need to describe the following things:

* Capture PlatformWorkedWith.

* Visualize as a Word cloud chart.

* Utilize Words, Size, Color fields of Word cloud chart.

* Include a proper Chart title.

```python
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
from wordcloud import STOPWORDS
```

```python
df_words = df[["Respondent" , "PlatformWorkedWith"]].groupby("PlatformWorkedWith").count().sort_values(by = "Respondent", ascending=False).reset_index()
df_words
```

```python
text = []
for i in range(17):
  m = df_words.loc[i, "Respondent"]
  A = df_words.loc[i, "PlatformWorkedWith"]
  for x in range(m):
    text.append(A)
#text
```

```python
texto = " ".join(i for i in text)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(texto)
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
