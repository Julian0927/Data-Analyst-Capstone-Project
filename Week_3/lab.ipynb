$(2)$ What is the median ConvertedComp before removing outliers?

```python
df["ConvertedComp"].median()
```

$ \ $

$(3)$ What is the median ConvertedComp after removing outliers?


```python
resumen = df["ConvertedComp"].describe()
diccionario = resumen.to_dict()
Q1 = diccionario['25%']
Q3 = diccionario['75%']
diccionario
```

```python
IQR = Q3-Q1
IQR 
```

```python
print(Q1 - 1.5 * IQR)
print(Q3 + 1.5 * IQR)
```

```python
df1= df[(df['ConvertedComp'] > (Q1 - 1.5 * IQR)) & (df['ConvertedComp'] < (Q3 + 1.5 * IQR))] 
sns.boxplot(x=df1["ConvertedComp"])
```

```python
df1['ConvertedComp'].median()
```
